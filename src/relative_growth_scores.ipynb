```python
# star score function
def star_score(num_appear, avg_revenue, avg_rating):
    return (num_appear + 0.17 * np.log(avg_revenue))**(np.log10(avg_rating))
```

```python
def minMaxScale(data, m = None): #min max scaling
    mini = min(data)
    maxi = max(data)
    
    if (m != None): mini = m
    
    return (data - mini)/(maxi - mini)
```

```python
def rel_avg_revs(revenues):
    k = 1
    avg_revs = []

    for i in range(len(revenues)):
        num_rev = revenues[:k]
        temp = np.mean(num_rev)
        avg_revs.append(temp)
        k += 1

    return avg_revs
```

```python
def rel_avg_ratings(ratings, num_votes):
    k = 1
    avg_ratings = []
    normalized_ratings = np.multiply(ratings, minMaxScale(np.log(num_votes), m = 0))
    
    for i in range(len(normalized_ratings)):
        num_ratings = normalized_ratings[:k]
        temp = np.mean(num_ratings)
        avg_ratings.append(temp)
        k += 1

    return avg_ratings
```

```python
def rel_star_score(revenues, ratings, num_votes):

    avg_rev = rel_avg_revs(revenues)
    avg_ratings = rel_avg_ratings(ratings, num_votes)

    rel_scores = []
    appearances = []
    for i in range(len(revenues)):
        scores_temp = star_score(i+1, avg_rev[i], avg_ratings[i])
        appearances_temp = i + 1
        rel_scores.append(scores_temp)
        appearances.append(appearances_temp)

    df = pd.DataFrame({'Appearances': appearances, 'Star_Scores':rel_scores})
    
    return df
```

```python
def score_growth(df, actor_name):
    plt.plot(df['Appearances'], df['Star_Scores'])
    plt.title(actor_name + 's Star Score Growth')
    plt.xlabel('Number of Appearances')
    plt.ylabel('Star Score per appearance')
    plt.show()
```
